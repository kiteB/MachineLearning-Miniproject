```python
wine_quality = wine['quality']
wine_data = wine.drop('quality', axis=1)
```

# 선형 회귀

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(wine_data, wine_quality)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

wine_predictions = lin_reg.predict(wine_data)
lin_mse = mean_squared_error(wine_quality, wine_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
from sklearn.metrics import mean_absolute_error

lin_mae = mean_absolute_error(wine_quality, wine_predictions)
lin_mae
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(random_state=42)
tree_reg.fit(wine_data, wine_quality)
```

```python
wine_predictions = tree_reg.predict(wine_data)
tree_mse = mean_squared_error(wine_quality, wine_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse
```

# 교차 검증

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_reg, wine_data, wine_quality, 
                         scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
    
display_scores(tree_rmse_scores)
```

```python
lin_scores = cross_val_score(lin_reg, wine_data, wine_quality, 
                             scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor(n_estimators=100,random_state=42)
forest_reg.fit(wine_data, wine_quality)
```

```python
wine_predictions = forest_reg.predict(wine_data)
forest_mse = mean_squared_error(wine_quality, wine_predictions)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

```python
from sklearn.model_selection import cross_val_score

forest_scores = cross_val_score(forest_reg, wine_data, wine_quality,
                                scoring="neg_mean_squared_error", cv=10)
forest_rmse_scores = np.sqrt(-forest_scores)
display_scores(forest_rmse_scores)
```

# 세부 튜닝

```python
from sklearn.model_selection import GridSearchCV

param_grid = [
    # try 12 (3×4) combinations of hyperparameters
    {'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},
    # then try 6 (2×3) combinations with bootstrap set as False
    {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},
  ]

forest_reg = RandomForestRegressor(random_state=42)
# train across 5 folds, that's a total of (12+6)*5=90 rounds of training 
grid_search = GridSearchCV(forest_reg, param_grid, cv=5,
                           scoring='neg_mean_squared_error',
                           return_train_score=True)
grid_search.fit(wine_data, wine_quality)
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

```python
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

```python
pd.DataFrame(grid_search.cv_results_)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_distribs = {
        'n_estimators': randint(low=1, high=200),
        'max_features': randint(low=1, high=8),
    }

forest_reg = RandomForestRegressor(random_state=42)
rnd_search = RandomizedSearchCV(forest_reg, param_distributions=param_distribs,
                                n_iter=10, cv=5, scoring='neg_mean_squared_error', random_state=42)
rnd_search.fit(wine_data, wine_quality)
```

```python
cvres = rnd_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

-> max_features: 5, n_estimators: 100일 때 최적의 솔루션!
