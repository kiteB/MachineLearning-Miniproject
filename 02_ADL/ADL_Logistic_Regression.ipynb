```python
import pandas as pd
import numpy as np
```

```python
ADL = pd.read_csv('train.csv')
```

```python
ADL.head()
```

# 데이터 분리

```python
ADL_labels = ADL['Activity']
ADL_labels
```

```python
ADL_data = ADL.drop(['Activity'], axis=1)
ADL_data
```

# 로지스틱 회귀

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(solver="liblinear",random_state=42)
```

```python
log_reg.fit(ADL_data, ADL_labels)
```

# 교차 검증

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(log_reg, ADL_data, ADL_labels, cv=10, scoring="accuracy")
```

```python
scores
```

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())

display_scores(scores)
```
