# ML miniproject #1

* EMNIST 데이터셋의 손글씨 숫자 이미지 분류

```python
import pandas as pd
import numpy as np

emnist = pd.read_csv('emnist-digits-train.csv')
```

```python
emnist.head()
```

```python
X = emnist.drop(['8'], axis=1) # 타겟 레이블을 제외한 데이터
X
```

```python
y = emnist['8'] # 타겟 레이블
y
```


# RandomForest

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=42)
```

```python
model.fit(X, y)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X, y, cv=10, scoring="accuracy")
```

```python
scores
```

```python
scores.mean() # 정확도 교차검증 평균
```
